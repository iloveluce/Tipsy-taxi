```python
# import relevant libraries
import pandas as pd
import numpy as np
import random
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# read in the data
taxi = pd.read_csv('master.csv')

# drop unnecessary columns
taxi.drop(['medallion_y','Unnamed: 0'], axis=1, inplace=True)

# convert times to datetime objects
taxi['pickup_datetime'] = [pd.to_datetime(a) for a in taxi['pickup_datetime']]
taxi['dropoff_datetime'] = [pd.to_datetime(a) for a in taxi['dropoff_datetime']]

# drop trips with no time
taxi = taxi[taxi.trip_time_in_secs > 10.0]

# generate a column for trip time in minutes
taxi['trip_time_min'] = taxi['trip_time_in_secs']/60.0

#eliminate outlier
taxi = taxi[(taxi['trip_time_min'] < 10000)]

# generate column for minutes since midnight
taxi['time'] = [a.hour*60 + a.minute for a in taxi['pickup_datetime']]

# generate a column for average speed of the trip
taxi['speed'] = taxi['trip_distance']/taxi['trip_time_min']

# drop trips with unreasonable speeds (in excess of 60mph)
taxi = taxi[taxi.speed < 1.0]

# generate a column for the day of the week
taxi['weekday'] = [a.weekday() for a in taxi['pickup_datetime']]

# generate a column to note whether it was a weekday or a weekend
taxi['weekend'] = [1 if (a == 5 or a == 6) else 0 for a in taxi['weekday']]

# drop trips with payment not cash or credit
cashcred = taxi[taxi.payment_type == ('CSH' or 'CRD')]

# rename medallion column
taxi = taxi.rename(columns={'medallion_x': 'medallion'})

# eliminate outliers
taxi = taxi[(taxi.pickup_latitude > 40.6) & (taxi.pickup_latitude < 40.9) & (taxi.pickup_longitude < -73.7) & (taxi.pickup_longitude > -74.1)]
taxi = taxi[(taxi.dropoff_latitude > 40.6) & (taxi.dropoff_latitude < 40.9) & (taxi.dropoff_longitude < -73.7) & (taxi.dropoff_longitude > -74.1)]
```

```python
# create a column for percent tip
taxi['percent_tip'] = taxi['tip_amount'] / taxi['fare_amount']

# create a column for time of day
taxi['time_of_day'] = ['morning' if (4 <= time.hour < 12) else 'afternoon' if (12 <= time.hour < 17) else 'evening' if (17 <= time.hour < 21) else 'night' for time in taxi['pickup_datetime']]

taxi.head()
```

```python
# see if tip percent is different for cash or card
grouped = taxi.groupby(['payment_type'], as_index=False)
mean = grouped.aggregate(np.mean)
#stdv = grouped.aggregate(np.std)

tip_percent = mean[['payment_type', 'percent_tip']]
print tip_percent
ind = np.arange(5)
heights = (tip_percent['percent_tip'][0], tip_percent['percent_tip'][1], tip_percent['percent_tip'][2], tip_percent['percent_tip'][3], tip_percent['percent_tip'][4])
plt.figure(figsize =(14,7))
plt.bar(ind, heights, width=0.70)
plt.ylabel('Tip Percentage')
plt.xticks(ind+0.70/2., (tip_percent['payment_type']))
plt.title('Average Tip Percentage for Each Payment Type')
plt.show()

# people may pay with card and be more likely to tip on only longer rides - check that next
```

```python
#see if tip percentage is different based on time of day
grouped = taxi.groupby(['time_of_day'], as_index=False)
mean = grouped.aggregate(np.mean)
#stdv = grouped.aggregate(np.std)

tip_percent = mean[['time_of_day', 'percent_tip']]
print tip_percent

ind = np.arange(4)
heights = (tip_percent['percent_tip'][2], tip_percent['percent_tip'][0], tip_percent['percent_tip'][1], tip_percent['percent_tip'][3])
plt.figure(figsize =(14,7))
plt.bar(ind, heights, width=0.70)
plt.ylabel('Tip Percentage')
plt.xticks(ind+0.70/2., (tip_percent['time_of_day'][2], tip_percent['time_of_day'][0], tip_percent['time_of_day'][1], tip_percent['time_of_day'][3]))
plt.title('Average Tip Percentage for Times of Day')
plt.show()
```

```python
# plot tip percentage vs. time of trip
plt.figure(figsize =(20,8))
plt.scatter(taxi['trip_time_min'], taxi['percent_tip'])
plt.xlabel('Trip Time in Minutes')
plt.ylabel('Percent Tip')
plt.xlim(-0.5,180)
plt.ylim(-0.1,20)
plt.title('Tip Percentage vs. Time of Trip')
```

```python
# plot tip percentage vs. speed of trip
plt.figure(figsize =(20,8))
plt.scatter(taxi['speed'], taxi['percent_tip'])
plt.xlabel('Speed of Ride')
plt.ylabel('Percent Tip')
plt.xlim(-0.005,1.0)
plt.ylim(-0.1,30)
plt.title('Tip Percentage vs. Speed')
```
