```python
# import relevant libraries
import pandas as pd
import numpy as np
import random
import matplotlib.pyplot as plt
import sklearn
import seaborn as sns

import sklearn.datasets
import sklearn.cross_validation
import sklearn.decomposition
import sklearn.grid_search
import sklearn.neighbors
import sklearn.metrics
from sklearn.ensemble import RandomForestClassifier
from sklearn.cross_validation import cross_val_score
from sklearn.feature_selection import SelectKBest, f_regression
from itertools import cycle

%matplotlib inline
```

```python
# read in the data
taxi = pd.read_csv('master_2.csv')

# drop unnecessary columns
taxi.drop(['Unnamed: 0'], axis=1, inplace=True)

# convert times to datetime objects
taxi['pickup_datetime'] = [pd.to_datetime(a) for a in taxi['pickup_datetime']]
taxi['dropoff_datetime'] = [pd.to_datetime(a) for a in taxi['dropoff_datetime']]

# generate a column for trip time in minutes
taxi['trip_time_min'] = taxi['trip_time_in_secs']/60.0

# generate column for minutes since midnight
taxi['time'] = [a.hour*60 + a.minute for a in taxi['pickup_datetime']]

# generate a column for hour of the day
taxi['hour'] = [a.hour for a in taxi['pickup_datetime']]

# generate a column for average speed of the trip
taxi['speed'] = taxi['trip_distance']/taxi['trip_time_min']

# drop trips with unreasonable speeds (in excess of 60mph)
taxi = taxi[taxi.speed < 1.0]

# generate a column for the day of the week
taxi['weekday'] = [a.weekday() for a in taxi['pickup_datetime']]

# generate a column to note whether it was a weekday or a weekend
taxi['weekend'] = [1 if (a == 5 or a == 6) else 0 for a in taxi['weekday']]

# generate a column to calculate the total fare
taxi['cost'] = taxi['total_amount'] - taxi['tip_amount']

# generate a table only for credit card values for tip data purposes
sub = taxi[taxi['payment_type'] == 'CRD']
```

```python
predictors = taxi[['time', 'passenger_count', 'trip_time_min', 'trip_distance', 'pickup_longitude', 'pickup_latitude', 
          'dropoff_longitude', 'dropoff_latitude', 'weekday']]
X = predictors.as_matrix()
Y = [round(a) for a in taxi['cost']]
Y = np.asarray(Y)

#split the data into a train and test set
X_train, X_test, Y_train, Y_test = sklearn.cross_validation.train_test_split(X, Y, test_size=0.33, random_state=99)

#print dimensions to ensure it worked properly
print X_train.shape, X_test.shape, Y_train.shape, Y_test.shape
```

```python
# define the values of k to check
ks = {'n_neighbors' : range(1,50)}

# define the estimator parameter
knn = sklearn.neighbors.KNeighborsClassifier()

# define the ten-fold cross validation
classifier = sklearn.grid_search.GridSearchCV(estimator = knn, param_grid = ks, cv = 10)

# fit the classifier to the data
classifier.fit(X_train,Y_train)

# find the optimal k value for the dataset
optimal_k = classifier.best_params_.values()[0]

# print the results
print 'Ten-fold cross validation determines that a k value of ' + str(optimal_k) + ' is optimal for the taxi dataset'
```

```python
# predict values on the test set
y_hat = classifier.predict(X_test)

# calculate accuracy of guesses
performance = sklearn.metrics.accuracy_score(Y_test, y_hat, normalize=True)

# print the results
print 'The tuned KNN classifier is ' + str(performance*100) + '% accurate on the taxi data set'
```

```python
# define the regression
regression = sklearn.linear_model.LinearRegression()

# set dummy values
lowest_error = 10000
best_features = []

# try different number of regressors
for i in range(1,10):
    
    # find best regressor(s)
    selector = SelectKBest(f_regression, k=i).fit(X_train,Y_train)
    features = np.where(selector.get_support())[0]
        
    # fit the regression
    regression.fit(X_train[:,features],Y_train)
    
    # calculate the mean square error
    square_error = np.mean((Y_test - regression.predict(X_test[:,features]))**2)
    
    # check to see if this performance is better
    if square_error < lowest_error:
        
        # if performane is better, save the current features
        lowest_error = square_error
        best_features = features

# fit the regression to whatever the best features were determined to be
regression.fit(X_train[:,best_features],Y_train)

# extract regression coefficients
coefs = regression.coef_
variables = predictors.columns

# initialize the sabermetric equation
sabermetric = 'FARE = '

# build the equation
for i in range(len(coefs)):
    plus = ' + ' if i != (len(coefs) - 1) else ''
    sabermetric += (str(coefs[i]) + '(' + str(variables[i]) + ')' + plus)

# print the final equation
print sabermetric

# calculate the mean square error
square_error = np.mean((Y_test - regression.predict(X_test[:,best_features]))**2)

print 'The regression has mean square error ' + str(square_error)
```
